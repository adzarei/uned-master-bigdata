## Ejercicios

- 1 Conectarse a la BD

- Acceder a la BD MyDatabase y comprobar que existe la colección MyCollection

- Poblar la colección MyCollection

Usar:
> for (var i=0;i<2000;i++){db.MyCollection.insert({‘supplierId’:i})}

- ¿Qué hemos hecho en la operación anterior?

> Se han insertado 2000 documentos en MyCollection con supplierId de 0 a 1999.

- ¿Por qué hemos utilizado el campo “supplierId” en concreto?

> Se ha usado ese campo porque es el Shard Key. <br/>
> Como supplierId es la Shard Key, los 2000 documentos se han repartido equitativamente entre los 3 shards.

> Como se puede ver en la celda de arriba, cada Shard ha alamacenado ,mas o menos, el mismo número de documentos.

- Realizar una consulta “find()” genérica sobre la colección y utilizar el método que se ha estudiado en la parte teórica para la monitorización de consultas, especificando la opción "executionStats". 

- ¿Qué resultados se obtienen?

> En la sección executionStages tenemos un nuevo stage que no se había visto antes: SHARD_MERGE.<br/>
> Hay una nueva sección llamada shards quer nos proporciona estadísticas de cada SHARD, por ejemplo elñ nº de documentos devueltos. <br/>
> La consulta ha accedido a los 3 shards, ha hecho un stage SHARDING_FILTER y por último un COLLSCAN.

- ¿Cuál es la explicación en lo relativo al sharding?

> Como los 2000 documentos se han repartido equitativamente entre cada SHARD ( ~ 666 documentos por SHARD ), el la consulta ha tenido que acceder a los 3 shards y recoger los documentos que contenía cada uno. <br/>
> En primer lugar se ha realizdo una operación de filtrado SHARDING_FILTER que descarta los documentos huerfanos de cada SHARD<br/>
> En segundo lugar se ha realizado un COLLSCAN para buscar, en cada shard, los documentos.
> por último, se han juntado los resultados en el stage SHARD_MERGE.
