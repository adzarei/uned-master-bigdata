```python
import requests

def getTweetsByConversationId(conv_id, limit=1000):
    url = "https://api.twitter.com/2/tweets/search/recent?query=conversation_id:{}&max_results={}&tweet.fields=id,text,created_at,author_id,conversation_id,in_reply_to_user_id&expansions=author_id,in_reply_to_user_id".format(conv_id, limit)
    print(url)
    
    payload={}
    headers = {
        'Authorization':'Bearer ' + bearer_token,
        'content-type':'application/json; charset=utf-8'
    }

    response = requests.request("GET", url, headers=headers, data=payload)

    json_response = json.loads(response.text)
    conversation_users = pd.DataFrame(json_response['includes']['users'])
    conversation_tweets = pd.DataFrame(json_response['data'])
    
    return (conversation_users, conversation_tweets)


def getRelationData(input_tweets):
    
    final_data = pd.DataFrame(columns = ['src_username','weight','trg_username','date'])
    
    for idx, tweet in input_tweets.iterrows():
        (users, tweets) = getTweetsByConversationId(tweet.conversation_id, 100)
    
        joined_data = pd.merge(tweets, users, left_on='author_id', right_on='id')[['username','in_reply_to_user_id', 'created_at']]\
            .rename(columns = {'username':'src_username', 'created_at': 'date'}, inplace = False)
        
        joined_data = pd.merge(joined_data, users, left_on='in_reply_to_user_id', right_on='id')[['src_username', 'date', 'username']]\
            .rename(columns = {'username':'trg_username'}, inplace = False)
        
        joined_data['weight'] = 1
        
        new_data = joined_data[['src_username','weight','trg_username','date']]
        final_data = pd.concat([final_data, new_data])
    
    return final_data
        
        
```
