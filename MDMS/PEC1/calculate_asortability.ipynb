# Notebook con el c√≥digo para calcular la asortavilidad desde una matriz de adyacencia.

```python
from scipy import stats
import pandas as pd
import os
```

```python
adj_mat_file = os.path.join('data','adj_mat_scenario_3_4.csv')
perason_corr_out_file = os.path.join('data','pearson_corr_scenario_3_4.csv')
```

```python
adj_mat = pd.read_csv(adj_mat_file, index_col=0)
adj_mat
```

```python
#Se crea una matriz 28x10 para contener las correlaciones
ady_pearson = pd.DataFrame(index=adj_mat.index, columns= adj_mat.index)
ady_pearson
```

```python
matrix_size = adj_mat.shape[0]
# matrix_size = 1
for i in range(matrix_size):
    for j in range(matrix_size):
        results = stats.pearsonr(adj_mat.iloc[i], adj_mat.iloc[j])
        ady_pearson.iat[i,j] = results[0]
```

```python
ady_pearson.to_csv(perason_corr_out_file)
```
