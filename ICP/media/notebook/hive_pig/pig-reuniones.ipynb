Reúne los datos de los clientes y del país al que pertenecen.

```python
%%writefile test.pig
clientes = load 'clients.csv' using PigStorage(',') as (nombrecliente:chararray, 
                                                        valoracion:chararray, 
                                                        codigopais:chararray);
paises = load 'countries.csv' using PigStorage(',') as (nombrepais:chararray, 
                                                        codigo:chararray);
reunion = join clientes by codigopais, paises by codigo;

reunionfiltrada = foreach reunion generate nombrecliente, valoracion, nombrepais;
dump reunionfiltrada;
```

```python
! pig -f test.pig
```

Muestra los nombres de todos los países y los datos de los clientes de cada país. 

```python
%%writefile test.pig
clientes = load 'clients.csv' using PigStorage(',') as (nombrecliente:chararray, 
                                                        valoracion:chararray, 
                                                        codigopais:chararray);

paises = load 'countries.csv' using PigStorage(',') as (nombrepais:chararray, 
                                                        codigo:chararray);

reunion = join clientes by codigopais right, paises by codigo;

reunionfiltrada = foreach reunion generate nombrecliente, valoracion, nombrepais;
dump reunionfiltrada;
```

```python
! pig -f test.pig
```

Muestra los datos de todos los clientes y de todos los países, reuniendo cada cliente con su país.

```python
%%writefile test.pig
clientes = load 'clients.csv' using PigStorage(',') as (nombrecliente:chararray, 
                                                        valoracion:chararray, 
                                                        codigopais:chararray);

paises = load 'countries.csv' using PigStorage(',') as (nombrepais:chararray, 
                                                        codigo:chararray);

reunion = join clientes by codigopais full, paises by codigo;

reunionfiltrada = foreach reunion generate nombrecliente, valoracion, nombrepais;
dump reunionfiltrada;
```
