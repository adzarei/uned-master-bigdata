## Primeros pasos con Pig

En este notebook veremos operaciones básicas con Pig. Cargaremos un esquema de Hive y veremos algunos de los operadores relacionales más importantes, como foreach, filter o group. También veremos cómo mostrar información por pantalla y almacenarla en HDFS.

```python
! mkdir -p pig-1erpasos
import os
os.chdir("pig-1erpasos")
! pwd
```

Cargaremos una tabla de Hive, y para cada fila, seleccionaremos la columna gsm19023 y almacenaremos estos datos en HDFS.

```python
%%writefile test.pig
datacleaninternal = load 'bioinformatica.datacleaninternal' using org.apache.hive.hcatalog.pig.HCatLoader();
dump datacleaninternal;
```

```python
! pig -f test.pig -x local
```

```python
%%writefile test.pig
datacleaninternal = load 'bioinformatica.datacleaninternal' using org.apache.hive.hcatalog.pig.HCatLoader();
gsm19023 = foreach datacleaninternal generate (gsm19023);
STORE gsm19023 INTO 'PIgsm19023c';
```

```python
! pig -f test.pig
```

```python
! hadoop fs -tail /user/root/PIgsm19023c/part-m-00000
```

También podemos ejecutar código en local sin utilizar Hadoop

Cargaremos una tabla de Hive, y para cada fila, seleccionaremos la columna gsm19023 y almacenaremos estos datos en otra tabla de Hive (que deberá estar creada priviamente).

```python
%%writefile test.hql
create table gsm19023tblC (
  gsm19023 FLOAT) 
ROW FORMAT DELIMITED FIELDS TERMINATED BY '\t';
```

```python
! beeline -u "jdbc:hive2://localhost:10000/bioinformatica" -f test.hql
```

```python
%%writefile test.pig
datacleaninternal = load 'bioinformatica.datacleaninternal' using org.apache.hive.hcatalog.pig.HCatLoader();
gsm19023 = foreach datacleaninternal generate (gsm19023);
STORE gsm19023 INTO 'bioinformatica.gsm19023tblC' using org.apache.hive.hcatalog.pig.HCatStorer();
```

```python
! pig -f test.pig
```

```python
! beeline -u "jdbc:hive2://localhost:10000/bioinformatica" -e "show tables;"
```

```python
! beeline -u "jdbc:hive2://localhost:10000/bioinformatica" -e "select * from gsm19023tblc;"
```

Ahora filtraremos las filas y nos quedaremos con aquellas cuyo campo gsm19023 se encuentre entre 100 y 200.

```python
%%writefile test.pig
datacleaninternal = load 'bioinformatica.datacleaninternal' using org.apache.hive.hcatalog.pig.HCatLoader();
gsm19023 = filter datacleaninternal by  gsm19023 > 100 and gsm19023 < 200;
dump gsm19023;
```

```python
! pig -f test.pig -x local
```

```python
%%writefile test.pig
datacleaninternal = load 'bioinformatica.datacleaninternal' using org.apache.hive.hcatalog.pig.HCatLoader();
gsm19023 = filter datacleaninternal by  gsm19023 > 100 and gsm19023 < 200;
STORE gsm19023 INTO 'PIgsm19023B-100200';
```

```python
! pig -f test.pig
```

```python
! hadoop fs -tail /user/root/PIgsm19023B-100200/part-m-00000
```

Ahora nos quedamos con las filas cuyo genetitle valga stomatin

```python
%%writefile test.pig
datacleaninternal = load 'bioinformatica.datacleaninternal' using org.apache.hive.hcatalog.pig.HCatLoader();
getitle = filter datacleaninternal by genetitle == 'stomatin';
STORE getitle INTO 'pigstomatin';
```

```python
! pig -f test.pig
```

```python
! hadoop fs -cat /user/root/pigstomatin/part-m-00000
```

Vamos a agrupar datos en función de su valor en el campo genetitle

```python
%%writefile test.pig
datacleaninternal = load 'bioinformatica.datacleaninternal' using org.apache.hive.hcatalog.pig.HCatLoader();
data = group datacleaninternal by genetitle;
STORE data INTO 'PIGgenetitleC';
```

```python
! pig -f test.pig
```

```python
! hadoop fs -cat /user/root/PIGgenetitleC/part-r-00000 | head
```

Muestra el plan de ejecución de un comando

```python
%%writefile test.pig
datacleaninternal = load 'bioinformatica.datacleaninternal' using org.apache.hive.hcatalog.pig.HCatLoader();
data = group datacleaninternal by genetitle;
explain data;
```

```python
! pig -f test.pig
```
